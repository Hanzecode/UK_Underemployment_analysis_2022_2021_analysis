```python
import pandas as pd
```

```python
df = pd.read_stata("lmn_indresp.dta",convert_categoricals= False)
```

```python
columns_needed = [
    "pidp", "lmn_sex", "lmn_dvage", "lmn_hiqual_dv", "lmn_gor_dv",
    "lmn_jbstat", "lmn_jbsemp", "lmn_jbsoc00_cc", "lmn_jbhrs", "lmn_prefhrs",
    "lmn_payhr_dv", "lmn_fimngrs_dv", "lmn_jbsat", "lmn_jbsat_pay",
    "lmn_jbsat_hours", "lmn_scghq1_dv", "lmn_sf12mcs_dv", "lmn_disdif1",
    "lmn_hlstat", "lmn_jbterm1", "lmn_indinui_xw"

]

col = []
for c in columns_needed:
    if c in df.columns:
        col.append(c)
```

```python
print("lmn_prfhrs" in df.columns)
```

```python
data = df[col]
```

```python
data.rename(columns={
    "pidp": "person_id",
    "lmn_sex": "gender",
    "lmn_dvage": "age",
    "lmn_hiqual_dv": "education_level",
    "lmn_gor_dv": "region",
    "lmn_jbstat": "employment_status",
    "lmn_jbsemp": "self_employed",
    "lmn_jbsoc00_cc": "occupation_code",
    "lmn_jbhrs": "actual_hours",
    "lmn_prefhrs": "preferred_hours",
    "lmn_payhr_dv": "hourly_pay",
    "lmn_fimngrs_dv": "monthly_income",
    "lmn_jbsat": "job_satisfaction",
    "lmn_jbsat_pay": "pay_satisfaction",
    "lmn_jbsat_hours": "hours_satisfaction",
    "lmn_scghq1_dv": "ghq_score",
    "lmn_sf12mcs_dv": "mental_health_score",
    "lmn_disdif1": "disability_status",
    "lmn_hlstat": "health_status",
    "lmn_jbterm1": "contract_type",
    "lmn_indinui_xw": "weight_uk"
}, inplace=True)
```
