```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df2021_hhresp = pd.read_stata("klm_hhresp.dta", convert_categoricals= False)
df2021_indall = pd.read_stata("klm_indall.dta", convert_categoricals= False)
df2022_hhresp = pd.read_stata("lmn_hhresp.dta", convert_categoricals= False)
df2022_indall = pd.read_stata("lmn_indall.dta", convert_categoricals= False)
df2021_indall2 = pd.read_stata("klm_indall.dta", convert_categoricals= False)
df2022_indall2 = pd.read_stata("lmn_indall.dta", convert_categoricals= False)


```

```python
df2021_indall2['pidp']
```

```python
df2021_indall2 = df2021_indall2.rename(columns={
    "bornuk_dv": "immigrant_status",
    "klm_birthy": "birth_year",
    "pidp":"person_id"
    
})
df2022_indall2 = df2022_indall2.rename(columns={
    "bornuk_dv": "immigrant_status",
    "lmn_birthy": "birth_year",
    "pidp":"person_id"})

```

```python
df2021_indall2 = df2021_indall2[["person_id",'immigrant_status','birth_year']]
df2022_indall2 = df2022_indall2[["person_id",'immigrant_status','birth_year']]
```

```python
data_immigrant = pd.concat([df2021_indall2,df2022_indall2],axis=0)
```

```python
data_immigrant.isnull().sum()
```

```python
immigrant_status_labels = {
    1: "UK-born",
    2: "Migrant",
    -1: "don't know",
    -2: "refusal",
    -7: "proxy",
    -8: "inapplicable",
    -9: "missing"
}
birth_year_labels = {
    -1: "don't know",
    -2: "refusal",
    -7: "proxy",
    -8: "inapplicable",
    -9: "missing"
}


data_immigrant["birth_year"] = data_immigrant["birth_year"].replace(birth_year_labels)
data_immigrant["immigrant_status"] = data_immigrant["immigrant_status"].replace(immigrant_status_labels)

```

```python
data_immigrant.to_csv("immigrant_data.csv", index= False)
```
